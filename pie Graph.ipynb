```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x)
plt.show()
```

# Adding Label

```python
x=[10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

# Change color

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]
c=["y","b","g","r"]

plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

# Adding percentage

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%")
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%")
plt.show()
```

# Creating shadow

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True)
plt.show()
```

# Change Radius

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=0.5)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,labeldistance=0.8)
plt.show()
```

# Rotate cirle

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,startangle=180)
plt.show()
```

# Change font size

```python
#we use testprops to change the font size but we pass value in this as a form of dictonary
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15})
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False)
plt.show()
```

# Line width

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,wedgeprops={"linewidth":8})
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",wedgeprops={"linewidth":4,"width":1.5})
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False,wedgeprops={"linewidth":2,"edgecolor":"m"})
plt.show()
```

# Changecenter

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"m"},
        center=(6,8))
plt.show()
```

# Rotate labels

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False,wedgeprops={"linewidth":4},rotatelabels=True)
plt.show()
```

# Add title

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False,wedgeprops={"linewidth":4},rotatelabels=False
       )
plt.title("Wscube Tec",fontsize=20)
plt.legend()
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False,wedgeprops={"linewidth":4},rotatelabels=False
       )
plt.title("Wscube Tec",fontsize=20)
plt.legend(loc=3)
plt.show()
```

# Multiple Graph

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=0.5,colors=c)      

plt.show()
```

# Ring pie chart

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")      

plt.show()
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")

plt.gca().add_artist(cr)

plt.show()
```
