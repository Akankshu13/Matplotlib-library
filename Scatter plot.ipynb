# Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

plt.scatter(x,y)
plt.show()
```

# Adding title

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no)
plt.show()
```

# Adding same Color

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no,color="r")
plt.show()
```

# Adding different colors

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no,c=colors)
plt.show()
```

# Adding dot sizes

```python
#we use alpha for dot tranparency and the value of alpha btw 0-1
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=["r","y","g","b","r","g","r"]
sizes=[10,20,40,50,20,10,70]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no,c=colors,s=sizes,alpha=0.5)
plt.show()
```

```python
#we use marker to change the shape of dot
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=["r","y","g","b","r","g","r"]
sizes=[100,200,400,500,200,100,700]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no,c=colors,s=sizes,alpha=1,marker="*",edgecolor="g",linewidth=2)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=["r","y","g","b","r","g","r"]
sizes=[100,200,400,500,200,100,700]

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.scatter(day,no,c=colors,s=sizes,alpha=1)
plt.show()
```

```python
#we use cmap (colormap) for color and for colorgraph we use colorbar function and cmap k andar bhut sare clr functn ate hai
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=[10,29,30,29,56,20,30]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.colorbar()

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=[10,29,30,29,56,20,30]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day,no,c=colors,s=sizes,cmap="Pastel1")
plt.colorbar()

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

# Adding label

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=[10,29,30,29,56,20,30]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
t=plt.colorbar()
t.set_label("Colorbar",fontsize=18)

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```

# Multiple Graph

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[5,19,4,7,20,56,22]

colors=[10,29,30,29,56,20,30]
sizes=[100,200,400,500,200,100,700]

plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.scatter(day,no2,c=colors,s=sizes,cmap="BuPu")

t=plt.colorbar()
t.set_label("Colorbar",fontsize=18)

plt.title("Scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No",fontsize=20)

plt.show()
```
