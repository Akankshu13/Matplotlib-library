# Change shape

```python
#We shape the with the help of notch function and it only takes value as true or false
```

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,notch=True)

plt.show()
```

# Horizontal 

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,vert=False)

plt.show()
```

# Width

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,widths=0.6)

plt.show()
```

```python
#always put data in label as a list
```

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,labels=["Python"])



plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,labels=["Python"],patch_artist=True)



plt.show()
```

# Show mean

```python
x=[1,2,3,4,5,6,7,8,5]

plt.boxplot(x,labels=["Python"],showmeans=True)



plt.show()
```

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True)



plt.show()
```

# Join the whis to outer circle

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,whis=2)



plt.show()
```

# Change outer circle shape

```python
#we join the whis to outer circle
```

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,sym="g+")



plt.show()
```

# HIde outer circle

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,sym=" ")



plt.show()
```

# Changing box and cap color

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"))



plt.show()
```

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"))



plt.show()
```

# Change circle color

```python
x=[40,50,60,70,80,50,120]

plt.boxplot(x,labels=["Python"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b")
           ,flierprops=dict(markeredgecolor="y"))



plt.show()
```

# Multiple plot

```python
x=[40,50,60,70,80,50,120]
x1=[40,50,60,70,80,50,120]

y=[x,x1]

plt.boxplot(y,labels=["python","c++"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b")
           ,flierprops=dict(markeredgecolor="y"))




plt.show()
```
