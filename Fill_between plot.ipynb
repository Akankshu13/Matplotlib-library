```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]

plt.plot(x,y,color="r")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x,y)
#plt.legend()

plt.title("python",fontsize=20)
plt.show()

```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]

plt.plot(x,y,color="r")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x=[2,4],y1=2,y2=4)
#plt.legend()

plt.title("python",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]

plt.plot(x,y,color="r")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x,y,color="g")
#plt.legend()

plt.title("python",fontsize=20)
plt.show()
```

```python
#where function does not work as a list so we import numpy and create array
#jab koi specific kam btana hota hai ki yha ho ra hai tab is plot ka use krte hai
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,2,3,4,5,6,7])

plt.plot(x,y,color="r")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.fill_between(x,y,color="g",where=(x>2)&(x<=4),alpha=0.5)
#plt.legend()

plt.title("python",fontsize=20)
plt.show()
```
