# How to change Axis in Graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[3,5,2,6,1,7]

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,5,2,6,1,7]

plt.plot(x,y)

plt.xticks(x,labels=["c","c++","java","php","SQL","Phyton"])
plt.yticks(y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,5,2,6,1,7]

plt.plot(x,y)

#plt.xticks(x,labels=["c","c++","java","php","SQL","Phyton"])
#plt.yticks(y)

plt.xlim(0,10)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,5,2,6,1,7]

plt.plot(x,y)

#plt.xticks(x,labels=["c","c++","java","php","SQL","Phyton"])
#plt.yticks(y)

plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
#with the help of xlim function we put x axis limit
```

```python
x=[1,2,3,4,5,6]
y=[3,5,2,6,1,7]

plt.plot(x,y)

#plt.xticks(x,labels=["c","c++","java","php","SQL","Phyton"])
#plt.yticks(y)

#plt.xlim(0,10)

plt.axis([0,10,0,7])
plt.show()
```

```python
#with the help of axis function we decide both x and y axis limit
```
