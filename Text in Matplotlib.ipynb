# How to Add text in plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)
plt.show()
```

# Adding title

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

# Adding text to graph

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)

plt.text(3,6,"java",fontsize=10,style="italic")
plt.show()
```

# Put java into a box

```python
#we use bbox for this we pass the value in this as a dictonary
```

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)

plt.text(3,6,"java",fontsize=10,style="italic",bbox={"facecolor":"y"})
plt.show()
```

# Adding Arrow

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)

plt.text(3,6,"java",fontsize=10,style="italic",bbox={"facecolor":"y"})

plt.annotate("python",xy=(1,3),xytext=(1.5,4),arrowprops=dict(facecolor="black",shrink=100))
plt.show()
```

# Add legend

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)

plt.text(3,6,"java",fontsize=10,style="italic",bbox={"facecolor":"y"})

plt.annotate("python",xy=(1,3),xytext=(1.5,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend("up",loc=1)
plt.show()
```

# Legend Face color change

```python
x=[1,2,3,4,5]
y=[3,2,6,1,5]

plt.plot(x,y)

plt.title("Python",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)

plt.text(3,6,"java",fontsize=10,style="italic",bbox={"facecolor":"y"})

plt.annotate("python",xy=(1,3),xytext=(1.5,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend("up",loc=1,facecolor="r",edgecolor="y",framealpha=0.5,shadow=True)
plt.show()
```
