# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
plt.xlabel("languages")
plt.ylabel("No")
plt.title("wscube")
plt.bar(x,y)
plt.show()
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)


plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)
plt.bar(x,y,width=0.4,color="r")
plt.show()
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
c=["y","r","g","b"]

plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)
plt.bar(x,y,width=0.4,color=c,align="edge")
plt.show()
```

```python
#there are two parameter in align 1 center and 2nd edge
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
c=["y","r","g","b"]

plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)
plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor="r",linewidth=1,linestyle=":",alpha=0.5)
plt.show()
```

# Multibar graph in one graph

```python
#now both graph overlab each other so in next step we make them seperate
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
z=(20,40,60,50)
plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)
plt.bar(x,y,width=0.4,color="r",align="edge",label="popularity")
plt.bar(x,z,width=0.4,color="g",align="edge",label="popularity1")
plt.legend()
plt.show()
```

```python
#we use np.arange to create the array of x taki use hum as a array use kare graph individual k liye and we create p1 t tranfer the 
```

```python
#value of p to p1 in the form of list
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
z=(20,40,60,50)

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)

plt.bar(p,y,width,color="r",align="edge",label="popularity")
plt.bar(p1,z,width,color="g",align="edge",label="popularity1")

plt.legend()
plt.show()
```

```python
#we use plt.ticks java pythOn add karne k liye and we only pass 3 parameters in this function and we use rotation for tilt x axis value
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
z=(20,40,60,50)

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)

plt.bar(p,y,width,color="r",align="edge",label="popularity")
plt.bar(p1,z,width,color="g",align="edge",label="popularity1")

plt.xticks(p+width/2,x,rotation=5)

plt.legend()
plt.show()
```

```python
x=("python","c","C++","java")
y=(80,70,60,82)
z=(20,40,60,50)



plt.xlabel("languages",fontsize=20)
plt.ylabel("No",fontsize=20)
plt.title("wscube",fontsize=20)

plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="g",label="popularity1")

plt.legend()
plt.show()
```
