```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]


plt.stackplot(x,area,area1,area2)
plt.show()
```

# Adding label

```python
#there are multiple area so we create label as a list
```

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]
l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l)

plt.legend(loc=2)
plt.show()
```

# Changing color

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]
l=["area","area1","area2"]
c=["deeppink","hotpink","orchid"]

plt.stackplot(x,area,area1,area2,labels=l,colors=c)

plt.legend(loc=2)
plt.show()
```

# Add title

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]
l=["area","area1","area2"]
c=["deeppink","hotpink","orchid"]

plt.stackplot(x,area,area1,area2,labels=l,colors=c,baseline="zero")

plt.title("Python",fontsize=20)
plt.legend(loc=2)
plt.show()
```

# Add axis

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]
l=["area","area1","area2"]
c=["deeppink","hotpink","orchid"]

plt.stackplot(x,area,area1,area2,labels=l,colors=c,baseline="zero")

plt.title("Python",fontsize=20)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```

# Add grid

```python
x=[1,2,3,4,5]
area=[5,2,6,4,8]
area1=[4,3,7,2,8]
area2=[3,6,4,6,2]
l=["area","area1","area2"]
c=["deeppink","hotpink","orchid"]

plt.stackplot(x,area,area1,area2,labels=l,colors=c,baseline="zero")

plt.title("Python",fontsize=20)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid()
plt.legend(loc=2)
plt.show()
```
