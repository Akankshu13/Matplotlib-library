```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,5,6,7,8]
y=[4,5,7,3,2,5]

plt.grid()
plt.step(x,y)
plt.show()
```

# Adding title and x&y label

```python
x=[2,4,5,6,7,8]
y=[4,5,7,3,2,5]

plt.grid()
plt.step(x,y)

plt.title("Python",fontsize=20)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Adding Marker

```python
x=[2,4,5,6,7,8]
y=[4,5,7,3,2,5]

plt.grid()
plt.step(x,y,marker="o")

plt.title("Python",fontsize=20)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Adding marker size and line color"

```python
x=[2,4,5,6,7,8]
y=[4,5,7,3,2,5]

plt.grid()
plt.step(x,y,color="r",marker="o",ms=15)

plt.title("Python",fontsize=20)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Change marker circle color

```python
x=[2,4,5,6,7,8]
y=[4,5,7,3,2,5]

plt.grid()
plt.step(x,y,color="r",marker="o",ms=15,mfc="g",label="python")

plt.title("Python",fontsize=20)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```
