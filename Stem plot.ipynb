```python
import matplotlib.pyplot as plt
```

```python
#we use stem plot to show signals
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y)

plt.show()
```

# Change line shape

```python
#we change the style of line
#solid line= "-"
#Dashed line="--"
#Dash dot line="-."
#Dotted line=":"
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt="-.")

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,markerfmt="r+")

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,markerfmt="ro")

plt.show()
```

# Change bottom value

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,bottom=2)

plt.show()
```

# Bottom line color change

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,basefmt="g")

plt.show()
```

# Add label

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,label="Python")

plt.legend()
plt.show()
```

# Change to Horizontal

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,orientation="horizontal")
plt.ylabel("Days",fontsize=20)
plt.xlabel("Numbers",fontsize=20)
plt.show()
```
